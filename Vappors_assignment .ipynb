```python
import pandas as pd

col_list = ["SysDt", "Gen_Voltage","Fuel_Ltrs"]
df = pd.read_csv("march_MT1080.csv", usecols=col_list)

#print (df)

#Convert Data Type of Date Column
df['SysDT'] = df['SysDt'].astype('datetime64[ns]')


#Adding Active and Inactive Status, Based on Generator Voltage
df = df.assign(Gen_Status='Inactive')

df.loc[df['Gen_Voltage'] > 100, 'Gen_Status'] = 'Active'

#print (df)
#fuel diiference, consecutive entries

df['Fuel_Difference'] = df['Fuel_Ltrs'].diff(-1)

print (df)
df.to_csv('full_data.csv', index=False)
#filtering active and inactive generator status

filt=(df['Gen_Status'] == 'Active')
#printing all active sgen status entries
print (df [filt])
df.to_csv('active_data.csv', index=False)
#print all inactive gen status entries
filt2= (df['Gen_Status'] == 'Inactive')

print (df [filt2])

df.to_csv('Inactive_data.csv', index=False)
```
